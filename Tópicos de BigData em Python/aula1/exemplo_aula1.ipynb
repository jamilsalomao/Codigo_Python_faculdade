```python
#imports
import pandas as pd
```

```python
df: pd.DataFrame = pd.read_csv(
    "dados_importacao.CSV",
    sep=";",
    encoding="cp1252",
    thousands=".",
    decimal=","
)
```

```python
df.head(5)
df.tail(5)
```

```python
df.columns = ["codigo", "data", "visita_otif", "ontime", "infull", "otif"]
```

```python
for column in df.columns:
    print(column, end=" ")

print("\n")

df.info()
```

```python
# sobre os dados do dataframe
df.describe()
```

```python
df["data"] = pd.to_datetime(df["data"], dayfirst=True)

df.dtypes
```

```python
df[["rev", "cod_cliente", "data_visita", "mapa"]] = df["visita_otif"].str.split("-", expand=True)
```

```python
df.drop("visita_otif", axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df["data_visita"] = pd.to_datetime(df["data_visita"], dayfirst=True)

df.head(5)
```

```python
troca: dict[str, int] = {"Sim": 1, "Nao": 0}
df["ontime"] = df["ontime"].map(troca)
df["infull"] = df["infull"].map(troca)
df["otif"] = df["otif"].map(troca)
```

```python
df.head()
```

```python
df["cod_cliente"] = df["cod_cliente"].astype('int64')
```
